```python
import PyPDF2

def get_page_number_from_index(pdf_path, char_index):
    # Otvaramo PDF fajl
    with open(pdf_path, 'rb') as f:
        reader = PyPDF2.PdfReader(f)
        total_chars = 0

        # Iteriramo kroz sve stranice da bismo izračunali kumulativni broj karaktera
        for page_num, page in enumerate(reader.pages):
            text = page.extract_text()
            page_length = len(text)
            total_chars += page_length

            # Ako je char_index unutar trenutne stranice, vraćamo broj stranice
            if total_chars >= char_index:
                return page_num + 1  # PDF stranice počinju od 1

    return None  # Ako indeks nije pronađen

# Primer kako pozvati funkciju za određeni indeks karaktera
pdf_path = '/Users/petarvukovic/Applications/apostol/back/files/bd-universe.pdf'
start_index = 4524

page_number = get_page_number_from_index(pdf_path, start_index)
print(f"Character index {start_index} is approximately on page {page_number}")
```
